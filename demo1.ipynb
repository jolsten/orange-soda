```python
from orangesoda.sources import OnesFrameSource, CycleFrameSource
from orangesoda.processor import NoOp, FrameFilter, CheckSingleByte
from orangesoda.sinks import StdoutPrinter
```

```python
source = CycleFrameSource(name='cycle-source', frame_size=64, num_frames=50, values=range(4), c_time="2000-01-01T00:00:00")

filter = FrameFilter(name='ffid', pass_func=CheckSingleByte(byte=0, value=3))
source.bind_output_to(filter)

unfiltered_sink = StdoutPrinter(name='stdout')
source.bind_output_to(unfiltered_sink)

filtered_sink = StdoutPrinter(name='stdout-filtered')
filter.bind_output_to(filtered_sink)

source.process()
filter.process()
```
