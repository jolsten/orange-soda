## Processing Flow Diagram

Let's make two simple processing chains from one data source.

The **source** outputs rows of data with values cycling through 0-3.

The first flow goes straight to a data **sink** that prints to STDOUT exactly what is generated by the **source**.

The second flow goes through a **filter** that only passes rows where byte 0 equals 2, then is printed by a second **sink**.

See diagram below:

```python
%%mermaid_magic

graph LR;
    source[(source)]
    filter[\filter/]
    sink1[unfiltered_sink]
    sink2[filtered_sink]
    source --> filter;
    source --> sink1;
    filter --> sink2;
```

```python
source = CycleFrameSource(name='cycle-source', frame_size=64, num_frames=24, values=range(4), c_time="2000-01-01T00:00:00")
unfiltered_sink = FramePrinter(name='stdout')

# Bind one block output Queue to the next block's input Queue
source.bind_output_to(unfiltered_sink)
```

```python
filter = FrameFilter(name='ffid', pass_func=CheckSingleByte(byte=0, value=2))
filtered_sink = FramePrinter(name='stdout-filtered')

# Alternatively, connect inputs with outputs using the __truediv__ operator
source / filter / filtered_sink
```

```python
# Process all the data in the blocks (must be done in order)
source.process()
filter.process()
```

```python
# The source provides a bunch of rows with cycling values for test purposes
unfiltered_sink.process()
```

```python
# The filter only passes rows that pass the filter criteria
filtered_sink.process()
```
